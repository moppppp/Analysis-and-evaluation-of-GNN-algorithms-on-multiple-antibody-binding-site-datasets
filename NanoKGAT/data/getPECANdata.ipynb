```python
import requests
import pandas as pd
import os

# 读取CSV文件（假设第一列名为'PDB_ID'）
df = pd.read_csv('/public/home/wangxin5/NanoKGAT/NanoKGAT-main/NanoKGAT-main/NanoKGAT/data/PECAN/train_set.csv')
pdb_ids = df.iloc[:, 0].tolist()  # 提取第一列所有PDB ID

# 创建保存目录
os.makedirs('pdb_filestrain', exist_ok=True)

# 遍历下载
for pdb_id in pdb_ids:
    url = f"https://files.rcsb.org/download/{pdb_id.upper()}.pdb"
    try:
        response = requests.get(url)
        response.raise_for_status()  # 检查请求是否成功
        with open(f"pdb_files/{pdb_id}.pdb", 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {pdb_id}.pdb")
    except Exception as e:
        print(f"Failed to download {pdb_id}: {e}")
```
