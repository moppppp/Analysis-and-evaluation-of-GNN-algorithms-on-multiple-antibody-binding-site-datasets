```python
import fnmatch
import re
from pathlib import Path
import pandas as pd
import shutil
```

```python


# 1. 定义源文件夹和目标文件夹路径
source_folder = Path('/home/chunhaixue/mxp/nanptopem/Nanotope-main/Nanotope-main/raw')
destination_folder = Path('/home/chunhaixue/mxp/nanptopem/Nanotope-main/Nanotope-main/PDB/train0218')
destination_folder.mkdir(parents=True, exist_ok=True)

# 2. 读取包含文件名的 DataFrame
val = pd.read_parquet("/home/chunhaixue/mxp/nanptopem//Nanotope-main//Nanotope-main//Nanotope/data/parquet/parquet0218/split1.parquet")
filenames = val['PDB'].dropna().unique().tolist()

# 3. 为每个文件名添加 .pdb 后缀
filenames_with_ext = [f"{name}.pdb" for name in filenames]
print(filenames_with_ext)
# 4. 查找匹配的文件（包含 .pdb 后缀）
matched_files = []
for file in source_folder.iterdir():
    if file.is_file():
        for pattern in filenames_with_ext:
            if fnmatch.fnmatch(file.name, pattern):
                matched_files.append(file)
                break

print(f"找到 {len(matched_files)} 个匹配的文件。")

# 5. 复制匹配的文件到目标文件夹
for file in matched_files:
    try:
        shutil.copy2(file, destination_folder / file.name)
        print(f"已复制: {file.name}")
    except Exception as e:
        print(f"复制文件 {file.name} 时出错: {e}")
```
